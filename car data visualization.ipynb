```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned car data.csv')
```

```python
df.head()
```

```python
df.drop(columns="Unnamed: 0",inplace=True)
df.info()
```

```python
fig, ax = plt.subplots(figsize=(15,6),ncols=3,dpi=400)
sns.countplot(data=df,y='company',ax=ax[0])
sns.countplot(data=df,y='year',ax=ax[1])
sns.countplot(data=df,y='fuel_type',ax=ax[2])
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(15,6),ncols=2,dpi=200)
sns.histplot(data=df,x='kms_driven',kde=True,bins=25,ax=ax[0])
sns.histplot(data=df,x='Price',kde=True,bins=25,ax=ax[1])
```

```python
sns.scatterplot(data=df,x="kms_driven",y='Price')
```

```python
plt.figure(figsize=(15,6),dpi=200)
sns.barplot(data=df,x="company",y='Price',hue='fuel_type')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,6),dpi=200)
sns.barplot(data=df,x="year",y='Price',hue='fuel_type')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df,x='fuel_type',y='Price')
```
